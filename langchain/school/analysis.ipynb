```python
import pandas as pd
```

```python
staar_2122 = pd.read_csv('austin-highschool-report-cards 2020-2022 - 2021-2022 STAAR.csv')

display(staar_2122)
```

```python
# In the text from the about pages, what subject mentioned does Asian have the highest score in 2022, for "Meet Grade Level"?

asian_scores = staar_2122[
    staar_2122['Section'].str.contains('Meets Grade Level')
].loc[
    staar_2122['Year'] == 2022
].loc[
    staar_2122['Subject'].isin(['Mathematics', 'Science', 'Social Studies']),
]['Asian'].str.strip('%').astype(int)

staar_2122.loc[asian_scores.idxmax()]['Subject']
```

```python
# What about lowest?

staar_2122.loc[asian_scores.idxmin()]['Subject']
```
