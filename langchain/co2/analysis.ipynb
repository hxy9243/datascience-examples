```python
import pandas

gdp_file = 'API_NY.GDP.MKTP.CD_DS2_en_csv_v2_5454986.csv'
co2_file = 'API_EN.ATM.CO2E.KT_DS2_en_csv_v2_5455005.csv'

gdp = pandas.read_csv(gdp_file, skiprows=4)
co2 = pandas.read_csv(co2_file, skiprows=4)
```

```python
gdp.head()
```

```python
co2.head()
```

```python
co2.to_csv('co2.csv')
gdp.to_csv('gdp.csv')
```

```python
import matplotlib.pyplot as plt

def plot_co2(country):
    years_col = [str(i) for i in range(1960, 2023)]
    country_co2 = co2[co2['Country Name'] == country][years_col]

    plt.figure()
    plt.plot(country_co2.T)
    plt.xticks(rotation=90)
    plt.grid()
    plt.show()

plot_co2('United States')
```

```python
plot_co2('China')
```

```python
plot_co2('Japan')
```

```python
plot_co2('Brazil')
```
